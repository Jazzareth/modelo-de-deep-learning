```python
df_train.columns = ['label', 'title', 'text']
df_train.head()
```

```python
df_test.columns = ['label', 'title', 'text']
df_test.head()
```

```python
#cambiamos los valores de label de 1 a 0 y de 2 a 1
df_train['label'] = df_train['label'].map({1:0, 2:1})
df_test['label'] = df_test['label'].map({1:0, 2:1})
```

```python
#Concatenamos el titulo junto con el texto
def concat_columns(df, col1, col2, new_col):
    df[new_col] = df[col1].apply(str) + ' ' + df[col2].apply(str)
    df.drop(col2, axis = 1, inplace = True)
    return df

df_train = concat_columns(df_train, 'text', 'title', 'text')
df_test = concat_columns(df_test, 'text', 'title', 'text')
```

```python
#Limpiamos los datos eliminando caracteres raros
def limpiar_text(text):
    text = re.sub(r'[^A-Za-zÀ-ú ]+', '', text)
    #text = re.sub('book|one', '', text)
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_train['text'] = df_train['text'].apply(limpiar_text)
df_test['text'] = df_test['text'].apply(limpiar_text)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

StopWord: es el nombre que reciben las palabras sin significado como artículos, pronombres, preposiciones, etc. que son filtradas antes o después del procesamiento de datos en lenguaje natural.

```python

def remove_stopwords(texto):
    stop_words = set(stopwords.words('english'))
    tokens = nltk.word_tokenize(texto.lower())
    return " ".join([token for token in tokens if token not in stop_words])

df_train['text'] = df_train['text'].apply(remove_stopwords)
df_test['text'] = df_test['text'].apply(remove_stopwords)
```
